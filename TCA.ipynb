```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a cross-tab of counts
counts = pd.crosstab(df["SeniorCitizen"], df["Churn"])

# Calculate percentages
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(4, 4))

bottoms = [0] * len(percentages)
for churn_status in percentages.columns:
    ax.bar(
        percentages.index,
        percentages[churn_status],
        bottom=bottoms,
        label=churn_status
    )
    # Update bottoms for stacking
    bottoms = [i + j for i, j in zip(bottoms, percentages[churn_status])]

# Add percentage labels
for i, row in percentages.iterrows():
    cumulative = 0
    for churn_status in percentages.columns:
        value = row[churn_status]
        if value > 0:  # avoid labeling 0%
            ax.text(
                i,
                cumulative + value / 2,
                f"{value:.1f}%",
                ha="center",
                va="center",
                color="white",
                fontsize=9,
                fontweight="bold"
            )
        cumulative += value

# Formatting
ax.set_title("Churn by SeniorCitizen (%)")
ax.set_xlabel("SeniorCitizen")
ax.set_ylabel("Percentage")
ax.legend(title="Churn")
plt.tight_layout()
plt.show()
```

```python
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 12))
axes = axes.flatten()  # Flatten for easy iteration

for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], order=df[col].value_counts().index, hue ="Churn")
    axes[i].set_title(f'Count Plot {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Hide any unused subplots (in case len(columns) isn't a perfect grid)
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#The plots show churn distribution across services: customers without online security, tech support, or backups, and those using fiber-optic internet, churn more often. Most customers have phone service, streaming, and internet services, with churn rates lower among those with added protections.
